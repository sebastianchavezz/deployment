```python
with open('database.json') as f:
    df = pd.read_json(f)
    
```

```python
df = df.T
```

```python

#generating genres op basis van new_df!!!
gen = []
count = 0
for genress in df['genres'].values:
    games = []
    if type(genress)== list:
        for g in genress:
            for des in g:
                if des == 'description':
                    games.append(g[des])
    
    else:
        games.append('no data')
    
    gen.append(games)
   
#joining the data into 1 list instead of multi dim
_genre = []

for lijst in gen:
    s = ', '.join(lijst)
    _genre.append(s)

```

```python
#gettin prices in euros, only currencies of euros wil be saved.

_pricess = []
counter = 0
for price in df['price_overview']:
    if type(price) == dict:
        for i in price:
            if i == 'currency' and price[i] != 'EUR':
                _pricess.append(None)
                break
            else:
               if i == 'final':
                   _pricess.append(price[i]/100)
    
    else:
        _pricess.append(0)



    counter += 1 

                   

           
  

_pricess
```

```python
#getting the developers as a string
count =0
deve = []
for dev in df['developers']:

    if type(dev) == list:
        count += 1
        s = ','.join(dev)
        
    deve.append(s)
print(count)
```
